```python
str_input = "1 2 6 3"

def list_of_squares(input_str):
    """ 
    На вход подаётся одна строка с целыми числами, разделёнными пробелом.
    Нужно с помощью map() получить список квадратов этих чисел и вывести его.
    """
    numbers = map(int, input_str.split())
    return list(map(lambda x : x ** 2, numbers))

print(list_of_squares(str_input))
```

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

def sum_list1(lst1, lst2):
    sum = list(map(lambda x, y: x + y, lst1, lst2))
    return sum

def sum_list2(lst1,lst2):
    return [ x + y for x, y in zip(lst1,lst2) ]

print(f"Вариант 1: {sum_list1(list1,list2)}")
print(f"Вариант 2: {sum_list2(list1,list2)}")
```

```python
temps = [20, 25, 30, 15]      # температура
humids = [60, 70, 50, 80]     # влажность % 

def normalize_data(temps, humids):
    t_min, t_max = min(temps), max(temps)
    h_min, h_max = min(humids), max(humids)
    normalised = list(map(lambda t, h: ((t - t_min) / (t_max - t_min) if t_max > t_min else 0, (h - h_min) / (h_max - h_min) if h_max > h_min else 0), temps, humids))
    return normalised
print(normalize_data(temps, humids))
```

```python
list11 = [3, 5, 67, 34, 9]
list12 = [3, 1, 2, 240, 76]

min_pair = min(list11), min(list12)
max_pair = max(list11), max(list12)
print(f"Минимальная пара (температура, влажность): {min_pair}")
print(f"Максимальная пара (температура, влажность): {max_pair}")
```
