```python
""" Пересечение множеств. """

orders = {
    "client1": {"apple", "banana", "orange"},
    "client2": {"banana", "apple", "grape"}, 
    "client3": {"orange", "kiwi"}
}

orders2 = {
    "client1": {"apple", "banana", "orange"},
    "client2": {"banana", "apple", "grape"}, 
    "client3": {"orange", "kiwi", "apple"}
}

print(type(orders))

def common_fruits(orders_dict):
    """ Функция отбирающая общий фрукт во всех заказах. """
    return set.intersection(*orders_dict.values())
print(common_fruits(orders))
print(common_fruits(orders2))
```

```python
orders3 = {
    "client1": {"apple", "banana", "orange"},
    "client2": {"banana", "apple", "grape", "kiwi"}, 
    "client3": {"orange", "kiwi", "pineapple"}
}

def unique_client_items(orders, client_name):
    """ Функция определяющая уникальный фрукт у клиента,
        который есть только у него одного. """
    client_set = orders[client_name]
    print(f"{ client_name } заказ { client_set }")
    others = set()  # объединяем всех остальных
    for name, items in orders.items():
        if name != client_name:
            others |= items  # union
    return client_set - others

print(unique_client_items(orders3,"client3"))
```

```python
club_a = {"Анна", "Борис", "Вера"}
club_b = {"Борис", "Глеб", "Дима"} 
club_c = {"Вера", "Дима", "Борис"}

clubs = {
    "club_a": club_a,
    "club_b": club_b, 
    "club_c": club_c
}

def club_stats(clubs):
    """ 
    Функция которая возвращает кортеж из 3 множеств:

    - Только в 1 клубе (никто больше)
    - Ровно в 2 клубах
    - Во всех 3 клубах
    """
    a, b, c = clubs.values()

    only_a = a - (b | c)
    only_b = b - (a | c) 
    only_c = c - (a | b)
    only_once = only_a | only_b | only_c

    exactly_two = ((a & b) - c) | ((a & c) - b) | ((b & c) - a)
    all_three = a & b & c
    
    return only_once, exactly_two, all_three

print(club_stats(clubs))
    
```

Вопрос 4: Какие навыки есть только у одного кандидата (уникальные)?

```python
# Подсчитаем, сколько раз встречается каждый навык
from collections import Counter

all_skills_list = []
for skills in candidates.values():
    all_skills_list.extend(skills)

skill_counts = Counter(all_skills_list)

# Навыки, которые встречаются только 1 раз
unique_skills = {skill for skill, count in skill_counts.items() if count == 1}
print("Уникальные навыки:", unique_skills)
```

Варианты ответа на вопрос 4 без Counter

Вложенный цикл с проверкой

```python
candidates = {
    'Анна': {'python', 'pandas', 'sql', 'machine_learning'},
    'Борис': {'python', 'sql', 'tableau', 'statistics'},
    'Виктория': {'python', 'pandas', 'deep_learning', 'tensorflow'},
    'Глеб': {'sql', 'tableau', 'excel', 'statistics'}
}

unique_skills = set()

# Для каждого кандидата проверяем его навыки
for candidate, skills in candidates.items():
    for skill in skills:
        # Считаем, у скольких кандидатов есть этот навык
        count = 0
        for other_skills in candidates.values():
            if skill in other_skills:
                count += 1
        
        # Если только у одного кандидата
        if count == 1:
            unique_skills.add(skill)

print("Уникальные навыки:", unique_skills)
```

Использование словаря для подсчёта

```python
skill_count = {}

# Подсчитываем вручную
for skills in candidates.values():
    for skill in skills:
        if skill in skill_count:
            skill_count[skill] += 1
        else:
            skill_count[skill] = 1

# Выбираем навыки с количеством 1
unique_skills = {skill for skill, count in skill_count.items() if count == 1}
print("Уникальные навыки:", unique_skills)
```

Более элегантный — через множества

```python
unique_skills = set()

for candidate, skills in candidates.items():
    # Создаём множество навыков всех ОСТАЛЬНЫХ кандидатов
    other_skills = set()
    for other_candidate, other_candidate_skills in candidates.items():
        if other_candidate != candidate:
            other_skills |= other_candidate_skills  # объединяем
    
    # Навыки, которых нет у других кандидатов
    unique_for_candidate = skills - other_skills
    unique_skills |= unique_for_candidate

print("Уникальные навыки:", unique_skills)
```

Самый лаконичный (через генераторы)

```python
unique_skills = {
    skill 
    for skills in candidates.values() 
    for skill in skills 
    if sum(skill in s for s in candidates.values()) == 1
}

print("Уникальные навыки:", unique_skills)
```

Через defaultdict (альтернатива Counter)

```python
from collections import defaultdict

skill_count = defaultdict(int)

for skills in candidates.values():
    for skill in skills:
        skill_count[skill] += 1

unique_skills = {skill for skill, count in skill_count.items() if count == 1}
print("Уникальные навыки:", unique_skills)
```
