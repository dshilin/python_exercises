```python
lst_0 = []
list_to_comp = [1, 1, 'g', 4, 4, 5, 'a', 'a', 5]
lst_compress = []

def compress(lst):
    """ Функция убирает подряд идущие символы в списке и оставляет только один. """
    for item in lst:
        if not lst_compress or item != lst_compress[-1]:
            lst_compress.append(item)
    return lst_compress

print(compress(list_to_comp))
print(compress(lst_0))
```

```python

def compress_str(s: str) -> []:
    """ 
    Сжимает строку по правилу RLE (Run-Length Encoding),
    но только для одинаковых подряд идущих символов.
    """
    if not s:
        return []
    i = 0
    res = []
    ln = len(s)
    while i < ln:
        curr_char = s[i]
        count = 1
        while i + 1 < ln and s[i + 1] == curr_char:
            count += 1
            i += 1
        res.append((curr_char, count))
        i += 1
    return res

print(compress_str("aaabbc")) 
print(compress_str("")) 
print(compress_str("hello")) 
    
        
```

```python
""" Списки списков. """

nested_list_1 = [[1,3,7,8], [23,5,4,3,0], [1,3,6,9,15,89]]

def merge_unique(lst) -> list:
    """ Функция принимает список списков, возвращает отсортированный список уникальных чисел из всех вложенных списков. """
    result = []
    for i in range(len(lst)):
        for j in range(len(lst[i])):
            if not any(lst[i][j] == existing for existing in result):
                result.append(lst[i][j])
    return sorted(result)

print(merge_unique(nested_list_1))
```

```python
def merged_unique_2(lst) -> list:
    unique = set()
    for sub in lst:
        for item in sub:
            unique.add(item)
    return sorted(unique)

print(merge_unique(nested_list_1))
```

```python
temps = (3, -1, 0, 4, 4, 2, -2)

def analyze_temps(temperature):
    temps_plus = 0
    for t in temperature:
        if t > 0:
            temps_plus = temps_plus + 1
    return min(temperature), max(temperature), temps_plus

print(analyze_temps(temps))
```

```python
import math


points = ((1, 2), (5, -3), (-2, 4), (3, 0), (3, 1))

def closest_to_origin(points) -> tuple:
    closest_dist = float('inf') 
    closest_point = None
    for point in points:
        x, y = point
        dist = math.sqrt(x**2 + y**2)
        if dist < closest_dist:
            closest_dist = dist
            closest_point = point
    return closest_point

print(closest_to_origin(points))
```

```python
list = __builtins__.list  # восстанавливаем оригинальную функцию
students = [
    ("Анна", 95, 88), 
    ("Борис", 78, 92), 
    ("Вера", 91, 89),
    ("Вера1", 91, 99),
    ("Вера2", 91, 99),
    ("Глеб", 85, 94)
]

def get_top_students(students, top_n):
    means = []
    names = []
    for st in students:
        names.append(st[0])
        means.append((st[1] + st[2])/2)
    students_mean = list(zip(names, means))
    return tuple(sorted(students_mean, key = lambda x : x[1], reverse=True)[:top_n])

print(get_top_students(students, 2))
```
